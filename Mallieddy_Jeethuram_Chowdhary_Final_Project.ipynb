```python
import pandas as pd

# Assuming df_bus is your DataFrame containing Yelp business data
# Replace 'df_bus' with the actual name of your DataFrame

# Firstly, ensure that there are no missing values in the relevant columns
df_bus_clean = df_bus.dropna(subset=['review_count', 'stars'])
df_bus_clean.head()
```

Answer 1:

The correlation coefficient between review_count and stars in your dataset is approximately 0.0324.

A value close to +1 indicates a strong positive correlation (more reviews correlate with higher ratings).

A value close to -1 indicates a strong negative correlation (more reviews correlate with lower ratings).

A value around 0 suggests no correlation.

while there is a slight positive relationship, it's so weak that it would not be advisable to draw strong conclusions or make business decisions based solely on this correlation.

```python
df_bus_clean = df_bus[df_bus['state'].str.isalpha()]
```

```python
# Assuming df_bus is your DataFrame
# Using regex to filter state column for values that consist of exactly two uppercase letters
valid_state_pattern = r'^[A-Z]{2}$'  # Regex pattern for two uppercase letters
df_bus_clean = df_bus[df_bus['state'].str.match(valid_state_pattern)]
```

```python
df_bus_clean.loc[:,"state"]
```

```python
import pandas as pd

# Define keywords that are indicative of food establishments
food_keywords = ['Restaurant', 'Cafe', 'Bakery', 'Diner', 'Bar', 'Food']

# Filtering the DataFrame for entries that are food establishments

# We use the 'str.contains' method and join the keywords with '|', which acts as an OR operator
df_food_establishments = df_bus_clean[df_bus_clean['categories'].str.contains('|'.join(food_keywords), na=False)]


```

```python
#  Group by City
grouped = df_food_establishments.groupby('city')
```

```python
# Initialize a list to store the winners
award_winners_list = []

# Iterate through each group (city)
for city, group in grouped:
    # Find the business with the highest rating
    max_rating = group['stars'].max()
    highest_rated_businesses = group[group['stars'] == max_rating]

    # Within the highest rated, find the one with the most reviews
    max_reviews = highest_rated_businesses['review_count'].max()
    winner = highest_rated_businesses[highest_rated_businesses['review_count'] == max_reviews]

    # Append the winner to the award_winners list
    award_winners_list.append(winner)

# Concatenate the list into a DataFrame
award_winners = pd.concat(award_winners_list)

# Display the winners
print(award_winners[['name', 'city', 'stars', 'review_count']])
```

There are instances where multiple establishments in the same city share the highest rating and review count. For example, in cities like Agincourt, there are different entries with the same or very similar high ratings but varying review counts.


The range of ratings and review counts varies significantly across cities. Some cities have top establishments with a 5.0-star rating and a substantial number of reviews, whereas others have lower ratings but still qualify as top due to the highest review counts in their specific locales.


The presence of multiple winners in some cities suggests a competitive food scene where several establishments manage to achieve high ratings and attract a considerable number of reviews.


```python
# Assuming 'award_winners' is the DataFrame with top establishments
plt.figure(figsize=(20, 10))
sns.barplot(data=award_winners, x='city', y='stars', hue='name')
plt.xticks(rotation=90)
plt.title('Top Rated Food Establishments in Each City')
plt.xlabel('City')
plt.ylabel('Rating (Stars)')
plt.legend(title='Establishment Name', loc='upper right')
plt.show()
```

```python
# Histogram of Ratings
plt.figure(figsize=(10, 6))
sns.histplot(award_winners['stars'], bins=10, kde=True)
plt.title('Distribution of Ratings Among Top Establishments')
plt.xlabel('Rating (Stars)')
plt.ylabel('Frequency')
plt.show()

# Histogram of Review Counts
plt.figure(figsize=(10, 6))
sns.histplot(award_winners['review_count'], bins=10, kde=True)
plt.title('Distribution of Review Counts Among Top Establishments')
plt.xlabel('Review Count')
plt.ylabel('Frequency')
plt.show()
```

```python
# Sentiment Analysis Function
def get_sentiment(text):
    if isinstance(text, str):  # Check if the text is a string
        analysis = TextBlob(text)
        # Classify as positive if polarity is > 0, else negative
        return 'positive' if analysis.sentiment.polarity > 0 else 'negative'
    else:
        return 'neutral'  # Return 'neutral' for non-string text
        
```

```python
sampled_reviews = df_review.sample(frac=0.1)  



# Apply sentiment analysis to the reviews
df_review['sentiment'] = df_review['text'].apply(get_sentiment)

 
```

```python
# Randomly select five businesses and extract their IDs
selected_businesses = df_bus['business_id'].sample(5).tolist()


# Filter reviews for these businesses
selected_reviews = df_review[df_review['business_id'].isin(selected_businesses)]

# Calculate the proportion of positive and negative reviews for each business
sentiment_proportion = selected_reviews.groupby(['business_id', 'sentiment']).size().unstack().fillna(0)
sentiment_proportion['total'] = sentiment_proportion.sum(axis=1)
sentiment_proportion['positive'] = sentiment_proportion['positive'] / sentiment_proportion['total']
sentiment_proportion['negative'] = sentiment_proportion['negative'] / sentiment_proportion['total']
```

```python

# Get average stars for these businesses
average_stars = df_bus[df_bus['business_id'].isin(selected_businesses)][['business_id', 'stars']]

# Merge with sentiment data
comparison_data = pd.merge(sentiment_proportion, average_stars, on='business_id')
```

```python
# Visualization
plt.figure(figsize=(10, 6))
x = np.arange(len(comparison_data))
width = 0.3

plt.bar(x - width/2, comparison_data['positive'], width, label='Positive Sentiment Proportion')
plt.bar(x + width/2, comparison_data['stars'], width, label='Average Stars Rating')

plt.xlabel('Business ID')
plt.ylabel('Proportion / Rating')
plt.title('Comparison of Positive Sentiment Proportion and Average Stars Rating')
plt.xticks(x, comparison_data['business_id'])
plt.legend()

plt.show()
```
